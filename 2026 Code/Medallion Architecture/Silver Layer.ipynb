# Let's assume that the we have a bronze layer and created the walmart table raw file in the workspace catalog and Walmart schema.

It is just the raw data that would come from s3 bucket or from pipeline...assuming

```python
%sql
CREATE OR REPLACE TABLE silver_walmart_sales
USING DELTA
AS
SELECT 
    Store,
    to_date(Date, 'yyyy-MM-dd') as Date,
    Weekly_Sales,
    Holiday_Flag,
    Temperature,
    Fuel_Price,
    CPI,
    Unemployment,
    year(to_date(Date, 'yyyy,MM-dd')) as Year,
    month(to_date(Date, 'yyyy,MM-dd')) as Month
FROM walmart.walmart
Where Weekly_Sales is NOT NULL
AND
    Weekly_Sales>0
```

```python
%sql
select * from silver_walmart_sales limit 50
```

#Doing a refinement step

```python
%sql
Update silver_walmart_sales
set
    Weekly_Sales=round(Weekly_Sales, 2),
    Temperature=round(Temperature, 2),
    Fuel_Price=round(Fuel_Price, 2),
    CPI=round(CPI, 2),
    Unemployment=round(Unemployment,2)
```
