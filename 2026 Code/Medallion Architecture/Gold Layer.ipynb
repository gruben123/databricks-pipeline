# Gold Layer and make it business ready

```python
%sql
CREATE OR REPLACE TABLE walmart.gold_sales_by_store
USING DELTA
AS
SELECT
    Store,
    round(sum(Weekly_Sales),2) as Total_Sales
FROM workspace.default.silver_walmart_sales
GROUP BY Store;
```

```python
%sql
select * from walmart.gold_sales_by_store
order by Total_Sales desc;
```

```python
%sql
CREATE OR REPLACE TABLE walmart.gold_holiday_impace
USING DELTA
AS
SELECT
    Holiday_Flag,
    round(sum(Weekly_Sales),2) as Total_Sales,
    round(avg(Weekly_Sales),2) as Avg_Sales
FROM workspace.default.silver_walmart_sales
GROUP BY Holiday_Flag;
```

```python
%sql
select * from walmart.gold_holiday_impace;
```

```python
%sql
CREATE OR REPLACE TABLE walmart.gold_monthly_sales
USING DELTA
AS
SELECT 
    Year,
    Month,
    round(sum(Weekly_Sales),2) as Total_Sales
FROM workspace.default.silver_walmart_sales
GROUP BY Year, Month
order by Year, Month;
```

```python
%sql
select * from walmart.gold_monthly_sales
```

```python
%sql
SELECT
    make_date(Year, Month, 1) as Sales_Date,
    Total_Sales
From walmart.gold_monthly_sales
ORDER BY sales_date
```
