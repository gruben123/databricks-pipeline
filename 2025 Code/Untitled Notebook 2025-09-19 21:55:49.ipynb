what are the functions of a groupBy

The groupBy function in PySpark is used to group the rows of a DataFrame based on the values of one or more columns. This is typically followed by an aggregation operation to compute summary statistics for each group. Here are some common functions used with groupBy:

1. **count()**: Counts the number of rows for each group.
2. **sum()**: Computes the sum of values for each group.
3. **avg()**: Computes the average of values for each group.
4. **max()**: Finds the maximum value for each group.
5. **min()**: Finds the minimum value for each group.
6. **agg()**: Allows for multiple aggregate functions to be applied to each group.
</p>
