```python
from pyspark.sql.functions import when

df=spark.read.table("customer")
df1=df.where(df.gender=="Female").limit(2)
display(df1)
display(df.schema)

#let's add a new column called salary and base it on age * 10000
df2=df1.withColumn("salary",df1.age*10000)

#let's add a new column called Seniority if age is less than 30, it's a Junior otherwise a Senior
df3=df2.withColumn("Seniority",when(df2.age<30,"Junior").otherwise("Senior"))
display(df3)
```

```python
df4=df3.withColumnRenamed("Seniority","Is_Seniority")
display(df4)
```

```python
df5=df4.drop("Is_Seniority")
display(df5)
display(df4)
```

```python
df6=df5.filter("customer_type=='Premium'")
display(df6)
```

```python
df7=df.na.fill({"email":"UNK", "age":0})
display(df7)
df8=df7.filter("age==0")
display(df8)
```

```python
from pyspark.sql.functions import desc, asc
df8=df7.orderBy(desc("age"), asc("registration_date"))
display(df8)
```

```python
from pyspark.sql.functions import col
df9=df.orderBy(col("email").asc_nulls_first())
display(df9)
```

```python
from pyspark.sql.functions import upper, ltrim, rtrim

df10=df.select(upper("email"))
display(df10)

df11=df.select(ltrim("country").alias("new_country"), rtrim("country"), ltrim("country"))
display(df11)
```

```python
df12=df.select("email", col("email").contains("gmail"))
display(df12)
```

```python
from pyspark.sql.functions import concat_ws
df13=df.select(concat_ws("@@", "age", "email", "country"))
display(df13)
```

```python
df1.write.mode("overwrite").option("header",True).csv("/FileStore/tables/customer.csv")
df15=spark.read.csv("/FileStore/tables/customer.csv", header=True)
display(df15)
```

```python
#you will have seperate folers for each country and the csv file will have all records from that country
df1.write.mode("overwrite").option("header",True).partitionBy("country").csv("/FileStore/tables/customer_partition")
df16=spark.read.csv("/FileStore/tables/customer_partition", header=True)
display(df16)
```
